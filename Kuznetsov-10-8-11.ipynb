```python
#Явный метод Эйлера
import math as m # подключаем библиотеки...
import numpy as np
import scipy as sci
import matplotlib.pyplot as plt
%matplotlib inline
A = np.array([[0, 1],[-100, -101]], dtype=float)# матрица якоби
start = np.array([2, -2])#краевые условия u0=(y(0)=2,z(0)=-2)
h = np.arange(0.020, 0.0202, 0.00002)# разные начальные h
def calc(h, x_max = 100.):#определяем функцию для каждого шага и нормируем всю область на 100
    N = int(x_max/h)#
    matrix = np.array([[1, 0], [0, 1]]) + h*A # матрица перехода явного метода Эйлера
    ans = np.zeros((N+1, 2))#вводим массив ответов
    ans[0] = start #Инициализация
    for idx in range(1, len(ans)):#проходим по узлам сетки
        ans[idx] = matrix.dot(ans[idx-1])#Явный метод Эйлера
    return (N+1, h, ans)#возвращение решений

for points in map(calc, h):#цикл по количеству разных шагов
    x = np.array([points[1]*i for i in range(0, points[0])])#h*i от 0 до N+1
    plt.plot(x, [v[0] for v in points[2]])#График(x,ans)
    plt.xlabel("$x$")#подпись оси х
    plt.ylabel("$y$")#подпись оси y
    plt.title(f"h ={points[1]}")#название
    plt.grid()#разметить сетку
    plt.show()#показать график
```

```python
откуда видно, что при h=0.02008 происходит резкое изменение u ближе к концу, что говорит о неустойчивости метода.
```
