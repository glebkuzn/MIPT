```python
#  
import math as m #  ...
import numpy as np
import scipy as sci
import matplotlib.pyplot as plt
%matplotlib inline
A = np.array([[0, 1],[-100, -101]], dtype=float)#  
start = np.array([2, -2])#  u0=(y(0)=2,z(0)=-2)
h = np.arange(0.020, 0.0202, 0.00002)#   h
def calc(h, x_max = 100.):#          100
    N = int(x_max/h)#
    matrix = np.array([[1, 0], [0, 1]]) + h*A #     
    ans = np.zeros((N+1, 2))#  
    ans[0] = start #
    for idx in range(1, len(ans)):#   
        ans[idx] = matrix.dot(ans[idx-1])#  
    return (N+1, h, ans)# 

for points in map(calc, h):#    
    x = np.array([points[1]*i for i in range(0, points[0])])#h*i  0  N+1
    plt.plot(x, [v[0] for v in points[2]])#(x,ans)
    plt.xlabel("$x$")#  
    plt.ylabel("$y$")#  y
    plt.title(f"h ={points[1]}")#
    plt.grid()# 
    plt.show()# 
```
