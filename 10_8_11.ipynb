```python
#Явный метод Эйлера
import math as m # подключаем библиотеки...
import numpy as np
end = 1 # конец сетки при t=1
h = np.arange(0.0001,0.01,0.0001) # разные начальные h
A = np.array([[0,1],[-100,-101]]) # матрица якоби
E = np.eye(2) # единичная матрица
u0 = np.array([2,-2]) #краевые условия u0=(y(0)=2,z(0)=-2)
#Явный метод Эйлера
for i in range(len(h)):
    u1 = u0
    nh = h[i] 
    Ai = E + h[i]*A
    while nh < end: # условие достижения последнего узла
        u2 = Ai.dot(u1) 
        nh += h[i]
        u1 = u2
    print('h=', h[i],'u=',u2) 
a = [2/m.exp(1),-2/m.exp(1)]
print('Teoretical: [x,y] =',a)
```

```python
откуда видно, что при h=0.001 происходит уменьшение u, что говорит о неустойчивости метода.
```
